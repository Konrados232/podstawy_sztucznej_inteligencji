Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))


# Plot outputs
plt.figure(figsize=(6,6))
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.show()
```

# Zadanie
wykonaj regresję z różnymi parametrami

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 1000))
regr_rf = RandomForestRegressor(max_depth=5, n_estimators =5, random_state=2)
y_rbf = regr_rf.fit(x, y)

# Plot outputs
plt.figure(figsize=(6,6))
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, regr_rf.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

# Zad
Wyznacz optymalne parametry w modelu RandomForestRegressor za pomocą GridSearchCV

```python
grid = GridSearchCV(estimator=make_pipeline(PolynomialFeatures(degree=2), RandomForestRegressor()),
                      param_grid={'polynomialfeatures__degree': [1, 2, 3, 4],
                                  'randomforestregressor__max_depth': [1, 2, 3, 4],
                                  'randomforestregressor__max_features': [1, 2, 3, 4],
                                  'randomforestregressor__n_estimators': [1, 2, 3, 4],
                                  'randomforestregressor__random_state': [1, 2, 3, 4]},
                      refit=True)

grid.fit(x, y)
grid.best_params_
```
