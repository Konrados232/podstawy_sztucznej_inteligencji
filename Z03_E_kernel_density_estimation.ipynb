```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from sklearn import cluster, datasets
```

```python
# normal distibution sample
mean = [0,0]
cov = [[3.8, 0.05], [0.05, 3.2]]

data = np.random.multivariate_normal(mean, cov, size=200)

x = data[:, 0]
y = data[:, 1]
X, Y = np.mgrid[-5:5:100j, -5:5:100j]

positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([x, y])
kernel = st.gaussian_kde(values)

f = np.reshape(kernel(positions).T, X.shape)

# data set sample
data_2, data_y_2 = datasets.make_circles(n_samples=200, factor=0.5, noise=0.05)

X_2, Y_2 = np.mgrid[-5:5:100j, -5:5:100j]

positions_2 = np.vstack([X.ravel(), Y.ravel()])
kernel_2 = st.gaussian_kde(data.T, bw_method= 0.2)

f_2 = np.reshape(kernel_2(positions_2).T, X_2.shape)

# plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

# contours
cfset = ax.contourf(X, Y, f, cmap='Blues')
cset = ax.contour(X, Y, f, colors='red')
#cfset = ax.contourf(X_2, Y_2, f_2, cmap='Blues')
#cset = ax.contour(X_2, Y_2, f_2, colors='red')


ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
plt.show()
```
