# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
n = [0, 0]
cov = [[1, 0], [0, 1]]
# cov = [[2, 0], [0, 1]]
# cov = [[1, 0], [0, 2]]

X = np.linspace(-5,5,500)
Y = np.linspace(-5,5,500)
X, Y = np.meshgrid(X,Y)

X1 = multivariate_normal(n, cov)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, X1.pdf(pos))
plt.show()
```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
n = [0, 0]
cov = [[1, 0], [0, 1]]
# cov = [[2, 0], [0, 1]]
# cov = [[1, 0], [0, 2]]

X = np.linspace(-5,5,500)
Y = np.linspace(-5,5,500)
X, Y = np.meshgrid(X,Y)

X2 = multivariate_normal(n, cov)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

fig = plt.figure(figsize=(5,5))
ax = plt.axes()
ax.contour(X, Y, X2.pdf(pos), 30)
plt.show()
```
